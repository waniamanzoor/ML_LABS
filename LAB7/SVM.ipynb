```python
from sklearn.svm import SVC
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Load a dataset
data = load_iris()
X = data.data
y = data.target

# Split into training and testing data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize and train the model with regularization
# C controls regularization. Try different values to see the effect
svc_model = SVC(kernel='linear', C=0.1, gamma=0.1)  # You can increase or decrease C
svc_model.fit(X_train, y_train)

# Predict on the test set
y_pred = svc_model.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy * 100:.2f}%")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Load the breast cancer dataset
data = load_breast_cancer()
X = data.data
y = data.target

# Split into training and testing data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize and train the SVM model with RBF kernel and gamma
svc_model = SVC(kernel='linear', C=1.0, gamma=0.1)  # You can adjust gamma and C
svc_model.fit(X_train, y_train)

# Predict on the test set
y_pred = svc_model.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy * 100:.2f}%")

# Visualize decision boundary (only using 2 features for simplicity)
# We'll use the first two features: 'mean radius' and 'mean texture'
X = X[:, :2]  # Reduce to only two features for visualization

# Split into training and testing data (again, for 2D visualization)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train the SVM again with the reduced features
svc_model.fit(X_train, y_train)

# Create a meshgrid for the plot
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.linspace(x_min, x_max, 100), np.linspace(y_min, y_max, 100))

# Predict for all points in the meshgrid
Z = svc_model.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Plot the decision boundary
plt.contourf(xx, yy, Z, alpha=0.8, cmap=plt.cm.coolwarm)

# Plot the training points
plt.scatter(X_train[:, 0], X_train[:, 1], c=y_train, s=30, edgecolors='k', cmap=plt.cm.coolwarm)
# Plot the test points
plt.scatter(X_test[:, 0], X_test[:, 1], c=y_test, s=30, edgecolors='k', marker='*', cmap=plt.cm.coolwarm)

# Add labels and title
plt.title('SVM Decision Boundary (Breast Cancer Dataset with RBF Kernel and Gamma)')
plt.xlabel('Mean Radius')
plt.ylabel('Mean Texture')
plt.show()
# Predict on the test set
y_pred = svc_model.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy * 100:.2f}%")
```
